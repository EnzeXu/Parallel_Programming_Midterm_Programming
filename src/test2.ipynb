```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')


def ground_truth(x0, x1):
    return x0**2 * x1 + x0 + 2 * x1

# def by_model(x0, x1):
#     return (1.4066*x1 + 0.3133)*np.exp(np.sin(x0))

def by_spl(x0, x1):
    return (1.4066*x1 + 0.3133)*np.exp(np.sin(x0))

x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)

X, Y = np.meshgrid(x, y)

ax.plot_surface(X, Y, ground_truth(X, Y), rstride=1, cstride=1,
                edgecolor='none')
ax.plot_surface(X, Y, by_spl(X, Y), rstride=1, cstride=1,
                edgecolor='none')
```

```python
import numpy as np
from svsr.brute_force import run_bfsr

task = "task_name"
grammars = ['A->A+A', 'A->A-A', 'A->A*A', 'A->A/A', 
            'A->x', 'A->C',
            'A->exp(A)', 'A->cos(x)', 'A->sin(x)']
equa = "X**2 + 2*X + 1"
X = np.random.uniform(*(-5, 5), 10)
Y = eval(equa)
train_sample = np.append(X, Y).reshape((2, 10))
run_bfsr(task, 
        grammars,
        train_sample,
        x_range = (-5, 5),
        node_limit = 8,
        hash_num = 5)
```

```python
def decorator_try(func: callable):
    def nfunc(*args, **kargs):
        print(f"before the call")
        func(*args, **kargs)
    return nfunc

@decorator_try
def hello_world():
    print("hello~")
    
@decorator_try
def print_my(s):
    print(s)

hello_world()
print_my("gg")
```
