```python
from srnn.utils import Trainer
import torch 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
FUNC_NAME = 'test2'
trainer = Trainer(
    model_mode='MLP',
    func_name=FUNC_NAME,
    batch_size=128,
    lr=0.0001,
    seed=0,
    cuda=0,
)
try:
    trainer.load()
except:
    trainer.fit(2000)
    trainer.load()
    trainer.model.cpu()
def neuro_eval(x: np.ndarray):              
    """assume the NN can eval for any x
    """
    if not isinstance(x, torch.Tensor):
        x = torch.tensor(x.T, dtype=torch.float32)
    rst2 = trainer.model(x).detach().numpy()
    return rst2

sample_freq = 100
x = np.linspace(-5, 5, sample_freq)
y = np.linspace(-5, 5, sample_freq)

X, Y = np.meshgrid(x, y)
def ground_truth(x0, x1):
    return x0**2 * x1 + x0 + 2 * x1
fig = plt.figure()
grou = ground_truth(X, Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, grou, rstride=1, cstride=1,
                edgecolor='none')
XY = np.append(X, Y).reshape((2, sample_freq * sample_freq))
neu = neuro_eval(XY).reshape((sample_freq, sample_freq))
print(neu.shape, X.shape)
ax.plot_surface(X, Y, neu, rstride=1, cstride=1,
                edgecolor='none')
plt.show()
tmp = neuro_eval(np.array([[4.74205896, 3.20109276], [-3.59141173, -3.59141173]], dtype=np.float64))
print(tmp)
print(ground_truth(4.74205896, -3.59141173))
print(ground_truth(3.20109276, -3.59141173))
delta = np.abs(neu - grou)
print(np.max(delta))

import seaborn as sns
```
